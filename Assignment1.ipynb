## Part 3:

__Z-score normalization to standardize income values:__

```python
income_z = (data["income"] - data["income"].mean()) / data["income"].std()
income_z.head(5)
```

__Add standardized income values into dataframe:__

```python
data["income_z"] = income_z
data.head(5)
```

## Part 4:

__Discretize age variable into "young", "mid-age", and "old" categories:__

```python
bin_labels = ["young", "mid-age", "old"]

age_bins = pd.qcut(data['age'], q = 3, labels = bin_labels)

age_bins.head(5)
```

__Add discretized age categories into dataframe:__

```python
data["age_disc"] = age_bins
data.head(5)
```

## Part 5:

__Min-max normalization to standardize age, income & children:__

```python
min_age = data["age"].min()
max_age = data["age"].max()
norm_age = (data["age"] - min_age) / (max_age-min_age)

min_inc = data["income"].min()
max_inc = data["income"].max()
norm_inc = (data["income"] - min_inc) / (max_inc-min_inc)

min_kids = data["children"].min()
max_kids = data["children"].max()
norm_kids = (data["children"] - min_kids) / (max_kids-min_kids)
```

__Replace age, income & children with normalized age, normalized income, & normalized children:__

```python
data["age"] = norm_age

data["income"] = norm_inc

data["children"] = norm_kids

data.head()
```

## Part 6:

__Convert categorical variables into dummy variables:__

```python
data_dummies = pd.get_dummies(data)

data_dummies.head()
```

__Write data_dummies into a CSV file:__

```python
data_dummies.to_csv("bank_numeric.csv")
```

Looking at the above correlation matrices, there appears to be a strong correlation between the variables age and income, with a 75.27% correlation (Pearson method).  Understandably, it follows that the discretized age variables age_disc_young (the first tercile of ages) and age_disc_old (the third tercile of ages) have correlations of -60% and 61.77%, respectively, with income (Pearson method).  There does not appear to be any other particularly strong correlations between the variables.

The two correlation methods used above were the Pearson and Spearman methods.  The Pearson correlation coefficient shows the linear relationship between variables, while the Spearman coefficient is based on the ranked values for each attribute.  Since the Spearman method is typically used to evaluate ordinal attributes rather than raw data, I primarily analyzed the above correlations pertaining to the first correlation matrix constructed using the Pearson method.
